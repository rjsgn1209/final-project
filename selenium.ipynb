### 무신사 크롤링 최종

```python
from selenium import webdriver

options = webdriver.ChromeOptions()
#       options.add_argument('--headless')
#       options.add_argument("--disable-extensions")
#       options.add_argument("disable-infobars")
#       options.add_argument("window-size=1920x1080")
options.add_argument("no-sandbox")
options.add_argument("disable-gpu")
options.add_argument("--lang=ko_KR")
options.add_argument('user-agent=Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36')


driver = webdriver.Chrome('data_files/chromedriver.exe', chrome_options=options)
driver.maximize_window()

for page in range(97, 100):
    start = time.time()
    df = pd.read_csv('data_files/img_info.csv', index_col='Unnamed: 0')
    df_temp = pd.DataFrame()
    print(f"{page} Page crawling...")
    url = f'https://www.musinsa.com/category/001?d_cat_cd=001&brand=&rate=&page_kind=search&list_kind=small&sort=pop&sub_sort=&page={page}&display_cnt=90&sale_goods=&group_sale=&kids=N&ex_soldout=&color=&price1=&price2=&exclusive_yn=&shoeSizeOption=&tags=&campaign_id=&timesale_yn=&q=&includeKeywords=&measure='
    driver.get(url)
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    product_num = len(driver.find_elements_by_css_selector('#searchList > li > div.li_inner > div.list_img > a '))
    for num in range(product_num):
        try:
            # ID가 myDynamicElement인 element가 로딩될 때 까지 10초 대기
            element = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "#searchList > li > div.li_inner > div.list_img > a"))
            )
            driver.find_elements_by_css_selector('#searchList > li > div.li_inner > div.list_img > a')[num].click()
        except TimeoutException:
            # 실패 시에는 에러메시지로 Time Out 출력
            print('Time Out')
            
        
            
        title = None
        img_url = None
        brand = None
        prduct_num = None
        season = None
        gender = None
        view_num = None
        sold_num = None
        like_num = None
        score = None
        review_num = None
        tag_li = None
        price =None
        
        try:
            # ID가 myDynamicElement인 element가 로딩될 때 까지 10초 대기
            element = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, ".product_title"))
            )
                     
            html = driver.page_source
            soup = BeautifulSoup(html, 'html.parser')
                     
            # 제목
            title = soup.select('.product_title')[0].text.replace('\n', '')
            title = re.sub('[^0-9a-zA-Zㄱ-힗 ]', '', title)

            # img_url
            if len(soup.select('.product-img img')) != 0:
                img_url = 'http:' + soup.select('.product-img img')[0]['src']
                urllib.request.urlretrieve(img_url, f"img/{title}.jpg")

            for li in soup.select('.product_article li'):
                try:
                    if li['style'] == 'display:none;':
                        pass
                    else:
                        raise KeyError
                except KeyError:
                    if len(li.select('.product_article_tit')) != 0:
                        info = li.select('.product_article_tit')[0].text.replace('\n', '').strip()
                        content = li.select('.product_article_contents')[0].text.replace('\n', '').strip()
                        if '/' in info:
                            for i in info.split('/'):
                                i = i.strip()

                                if '브랜드' in i:
                                    brand = content.split('/')[0]

                                elif '품번' in i:
                                    product_num = content.split('/')[1]

                                elif '시즌 ' in i:
                                    data = content.replace('\n', '').split()
                                    if len(data) == 1:
                                        season = None
                                    elif len(data) == 2:
                                        season = data[0]
                                    else:
                                        season = data[0] + ' ' + data[1]

                                elif '성별' in i:
                                    data = content.replace('\n', '').split()
                                    if len(data) == 1:
                                        gender = data[0]
                                    elif len(data) == 2:
                                        gender = data[1].replace('/', '')
                                    else:
                                        gender = data[2].replace('/', '')

                        if '조회수(1개월)' in info:
                            view_data = content.split('회')[0].strip()
                            unit = view_data[-1]
                            view_num = float(view_data[:-1])

                            if unit == '만':
                                view_num = int(view_num * 10000)
                            elif unit == '천':
                                view_num = int(view_num * 1000)
                            else :
                                view_num = int(view_num)

                        elif '누적판매(1년)' in info:
                            sold_data = content.split('개')[0].strip()
                            unit = sold_data[-1]
                            sold_num = float(sold_data[:-1])

                            if unit == '만':
                                sold_num = int(sold_num * 10000)
                            elif unit == '천':
                                sold_num = int(sold_num * 1000)
                            else :
                                sold_num = int(sold_num)

                        elif '좋아요' in info:
                            like_num = int(content.replace(',', ''))

                        elif '구매 후기' in info:
                            score = float(content.split('/')[0])
                            review_num  = int(re.sub(r'[^0-9]', '', content.split('/')[1]))

            # 태그
            if soup.select('li.article-tag-list'):
                data = soup.select('li.article-tag-list')[0]
                tags = data.select('.listItem')
                tag_li = []
                for tag in tags:
                    tag_li.append(tag.text.replace('#', ''))
            else:
                tag_li = None

            # 가격
            price = soup.select('.product_article_price')[0].text.replace(',', '').replace('\n', '').strip()
            if price[-1] == '원':
                price = price[:-1]
            price = int(price)

            # 데이터 프레임 추가
            dic = {'제목':title, 'img_url': img_url, '브랜드': brand, '품번': product_num, '시즌': season, '성별': gender, \
                   '조회수': view_num, '판매수': sold_num, '좋아요': like_num, '평점': score, '리뷰수': review_num, \
                   '태그': tag_li, '가격': price}
            df_temp = df_temp.append(dic, ignore_index=True)


            print('=' * 50)
            print(num, title)
        #     print(dic)
        #     print('브랜드 / 품번 ', brand, product_num)
        #     print('시즌 / 성별 ', season, gender)
        #     print('누저 조회수(1개월) ', view_num)
        #     print('누적 판매수(1년) ', sold_num)
        #     print('좋아요 ', like_num)
        #     print('구매후기 ', score, review_num)
        #     print(tag_li)
        #     print('구입가격', price)
        except TimeoutException:
            # 실패 시에는 에러메시지로 Time Out 출력
            print('Time Out')
        
        
        driver.back()
    df_save = df.append(df_temp, ignore_index=True)
    df_save.drop_duplicates('제목', inplace=True)
    df_save.reset_index(drop=True, inplace=True)
    df_save.to_csv('data_files/img_info.csv')
    end = time.time()

    print(f"End. {end - start:.5f} sec")
```

```python
titles = os.listdir('img_files/origin_pants_img/')
```

```python
max_file_num = len(titles)
```

```python
def auto_img_searching(input_text, start_num, end_num, step):
    data_dic = {}
    data_dic['step'] = step
    data_dic['input_text'] = input_text
    error_count = 0
    for num in range(start_num, end_num, step):
        start = time.time()
        data_dic['num'] = num
        with open('data_files/data.pickle', 'wb') as fw:
            pickle.dump(data_dic, fw)
        driver.find_element_by_css_selector('#kernellink').click()
        time.sleep(0.5)
        driver.find_element_by_css_selector('#restart_run_all > a').click()
        time.sleep(0.5)
        driver.find_element_by_css_selector('body > div.modal.fade.in > div > div > div.modal-footer > button.btn.btn-default.btn-sm.btn-danger').click()


        while True:
            time.sleep(3)
            try:
                driver.find_element_by_css_selector('#notebook-container > div:nth-child(11) > div.output_wrapper > div.output > div > div.output_subarea.output_text.output_stream.output_stdout > pre')
                print('Down Range :', num, '~', num+step)
                print("Time :", time.time() - start)
                error_count = 0
                break
            except:
                print('.', end='')
                error_count += 1
                if error_count > 10:
                    print()
                    print('*' * 50)
                    print(f"Error Break {num} ~ {num+step}")
                    print('*' * 50)
                    error_count = 0
                    break
```

```python
auto_img_searching('Slacks', 0, max_file_num, 180)
```
